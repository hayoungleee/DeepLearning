```
import numpy as np
import matplotlib.pyplot as plt
```

- 활성화 함수

```
def step_function(x):
  if x > 0 : return 1
  else: return 0
```

```
def sigmoid(x):
  return 1 / ( 1 + np.exp(-x))
```

```
# 항등함수
def identity_function(x):
  return x
```

```
def softmax(a):
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a
  return y
```

- 신경망 구현하기

```
# input 정의 //  1 x 2 
x=np.array([1.0,0.5])
# 행렬의 내적을 위해 shape 이 맞아야하기 때문에 정의시 조심해야한다.
# 가중치, 편향 정의
network = {}
# 2 x 3 
network['W1'] = np.array([
                          [0.1, 0.3, 0.5],
                          [0.2, 0.4, 0.6]
                          ])
# 1 x 3 
network['b1'] = np.array([0.1, 0.2, 0.3])

# 3 x 2 
network['W2'] = np.array([[0.1, 0.4], 
                          [0.2, 0.5], 
                          [0.3, 0.6]])
# 1 x 2 
network['b2'] = np.array([0.1, 0.2])

network['W3'] = np.array([[0.1, 0.3],
                          [0.2, 0.4]])
network['b3'] = np.array([0.1, 0.2])

# input -> 1층
a1 = np.dot(x,network['W1']) + network['b1']
z1 = sigmoid(a1)
# 1층 -> 2층
a2 = np.dot(z1,network['W2']) + network['b2']
z2 = sigmoid(a2)
# 2층 -> 3층
a3 = np.dot(z2,network['W3']) + network['b3']

# 3층 -> 활성화 함수 -> y.

y_identity = identity_function(a3)
y_softmax = softmax(a3)
```

```
print(y_identity)
print(y_softmax)
```
